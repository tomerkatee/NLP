```python
X = np.linspace(-10, 10, 20)

Y = 1/(1+np.exp(-X))
Yminus = 1/(1+np.exp(X))


plt.plot(X,Yminus)
plt.plot(X,Y)
plt.show()
```

The relation between (x) and (-x) is they are mirrored around the Y axis.

The derivative of (x) is:

$$
\frac{e^{-x}}{(1+e^{-x})^2}
$$
